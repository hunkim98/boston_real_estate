```python
data_pids = data_23["PID"].unique()

# Convert integers to strings and pad with zeros to 10 digits
data_pids_10 = [str(i).zfill(10) for i in data_pids]

data_pids_10_set = set(data_pids_10)

data_pids_10_set
```

```python
# check if all sales_pid_no_dash_set are included in data_pids_10_set

sales_pid_no_dash_in_data_pids = [i for i in sales_pid_no_dash_set if i not in data_pids_10_set]

print(sales_pid_no_dash_in_data_pids)

print(len(sales_pid_no_dash_in_data_pids))



```
