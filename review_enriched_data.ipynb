```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
data_folder = "data/2023"

file_name = "geocoded_merged_sales_fixed.csv"

data_path = f"{data_folder}/{file_name}"

data = pd.read_csv(data_path)



data[data["full_address"].isna()]


# check which do not have full_address


```

```python
# for rows that have full_address na we create the full address manually
for index, row in data[data["full_address"].isna()].iterrows():
    # only if street name is not na
    if row['street_name'] is not None:
        data.loc[index, "full_address"] = f"{row['street_name']} {row['street_no']} Boston MA"
    
data[data["x"].isna() | data["y"].isna()]


```

```python
data[data["x"].isna() | data["y"].isna()][["full_address", "parcel"]]
```

```python
from geocode_data import geocode_address

parcels_with_no_x_y = data[data["x"].isna() | data["y"].isna()]
# print(parcels_with_no_x_y["parcel"].values)

count = 0
for parcel in parcels_with_no_x_y["parcel"].values:
    # update the dataframe
    count += 1
    print(f"Geocoding parcel {parcel}")
    x, y = geocode_address(data[data["parcel"] == parcel]["full_address"].values[0])
    data.loc[data["parcel"] == parcel, "x"] = x
    data.loc[data["parcel"] == parcel, "y"] = y
    print(f"{count}/{len(parcels_with_no_x_y['parcel'].values)}")















```
